```python
# Importing libraries

import requests
import pandas as pd
import json
import time
from datetime import datetime, timedelta
```

```python
#Fetch Mutual fund Scheme codes

def get_all_schemes():
  #Fetch list of all mutual fund schemes
  response = requests.get("https://api.mfapi.in/mf")
  if response.status_code == 200:
    return response.json()
  return[]

print("Fetching scheme codes")
schemes = get_all_schemes()
```

```python
#checking if our code worked

print(f'total schemes found : {len(schemes)}')
```

SELECT SCHEMES AND MANUALLY & FETCH PRICE DATA FOR OUR SCHEMES (NAV)

```python
# Pick 9 large-cap / popular schemes (avoid overloading API)
# Manually select a good mix of fund houses
selected_schemes = [
    {'code': '119551', 'name': 'SBI Bluechip Fund - Growth'},
    {'code': '154321', 'name': 'Zerodha Nifty LargeMidcap250 Plus 8-13 Yr G-Sec 70:30 Index Fund - Direct Plan - Growth Option'},
    #{'code': '102885', 'name': 'HDFC Top 100 Fund - Growth'},
    {'code': '118499', 'name': 'Kotak Bluechip Fund - Growth'},
    {'code': '118960', 'name': 'Axis Bluechip Fund - Growth'},
    {'code': '119553', 'name': 'SBI Small Cap Fund - Growth'},
    {'code': '118669', 'name': 'HDFC Mid-Cap Opportunities Fund - Growth'},
    {'code': '120313', 'name': 'Nippon India Small Cap Fund - Growth'},
    {'code': '118523', 'name': 'SBI Contra Fund - Growth'},
]

# ==========================================================
# 2. FETCH NAV HISTORY FOR EACH SCHEME
# ==========================================================

def fetch_nav_history(scheme_code, scheme_name):
    """Fetch historical NAV for a scheme from MFAPI"""
    url = f"https://api.mfapi.in/mf/{scheme_code}"
    try:
        response = requests.get(url, timeout=10)
        if response.status_code == 200:
            data = response.json()
            meta = data.get('meta', {})
            nav_data = data.get('data', [])

            # Convert to DataFrame with meta info
            if nav_data:
                df = pd.DataFrame(nav_data)
                df['scheme_code'] = scheme_code
                df['scheme_name'] = scheme_name
                df['fund_house'] = meta.get('fund_house', '')
                df['scheme_type'] = meta.get('scheme_type', '')

                # Clean date format
                df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y', errors='coerce')
                df['nav'] = pd.to_numeric(df['nav'], errors='coerce')

                return df
        return pd.DataFrame()
    except Exception as e:
        print(f"Error fetching {scheme_name}: {e}")
        return pd.DataFrame()

# ==========================================================
# 3. COLLECT ALL DATA
# ==========================================================

all_data = []

print("\n" + "="*60)
print("Downloading NAV data for selected schemes...")
print("="*60)

for scheme in selected_schemes:
    print(f"\nFetching: {scheme['name']}...")
    df = fetch_nav_history(scheme['code'], scheme['name'])

    if not df.empty:
        all_data.append(df)
        print(f"  ✅ {len(df)} NAV records")
        print(f"  📅 {df['date'].min().date()} to {df['date'].max().date()}")
        print(f"  📊 NAV range: ₹{df['nav'].min():.2f} - ₹{df['nav'].max():.2f}")
    else:
        print(f"  ❌ No data")

    time.sleep(0.5)  # Be kind to the API

# Combine all data
if all_data:
    master_df = pd.concat(all_data, ignore_index=True)
    master_df = master_df.sort_values(['scheme_name', 'date'])
else:
    print("No data collected!")
    exit()

print("\n" + "="*60)
print("DATASET CREATED")
print("="*60)
print(f"Total records: {len(master_df)}")
print(f"Unique schemes: {master_df['scheme_name'].nunique()}")
print(f"Date range: {master_df['date'].min()} to {master_df['date'].max()}")
print(f"Columns: {list(master_df.columns)}")

# Save raw data
master_df.to_csv('mutual_fund_raw.csv', index=False)
print("✅ Saved: mutual_fund_raw.csv")
```

### The Problem with Accuracy for a Classification Model

As an example we are considering lets consider a very commonproblem : fraud detection

Lets say as a data scientist we created a basic model for fraud detection, and given that only about 1-2% transactions are fraud, lets say our model is REALLY Bad, it will still be 99% ACCURATE.

Thats why, for an imbalanced dataset, accuracy is not the right metric

### Precision and Recall

Precision : How accurately can our model identify an anomaly. Out of all the predicted frauds, how many of these were correctly identified? A high precision score means that when our model says that a transaction is fraud, it is usually correct. This metric is important when the cost of a false positive is high.

Recall : Of ALL the transactions that were ACTUALLY an anomaly, how many were caught by our model?A high recall score means that our model is catching most of the fraud transactions.

Often Precision and Recall are Sitting on a See-Saw in real life.
When a model has good precision its recall value might go down, and when a model has good recall, its precision value might go down.

F1 score : The harmonic mean of Precision and Recall which provides a single number that balances both metrics. Very useful when we dont have a clear prefernce between recall and precision or for comparison between different algorithms



```python
#Precision and Recall

from sklearn.metrics import precision_score, recall_score,f1_score

#Calulating Precision, Recall and F1 score

precision = precision_score(y_test,y_pred)

recall = recall_score(y_test,y_pred)

f1 = f1_score(y_test,y_pred)

print(f'Precision : {precision :.2%}')
print(f'Recall : {recall : .2%}')
print(f'F1 Score : {f1 : .2%}')
```

```python
%%writefile streamlit2.py

#my_1st_streamlit.py

import streamlit as st
import yfinance as yf
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import warnings
warnings.filterwarnings('ignore')

# ==========================================================
# PAGE CONFIGURATION
# ==========================================================

st.set_page_config(
    page_title="Stock Direction Predictor",
    page_icon="📈",
    layout="wide"
)

st.title("📈 Stock Market Direction Predictor")
st.markdown("This application predicts whether a stock will close **higher** or **lower** than the previous day.")

# ==========================================================
# SIDEBAR - USER INPUTS
# ==========================================================

st.sidebar.title("Input Parameters")

# Stock selection
ticker = st.sidebar.text_input(
    "Stock Ticker (Yahoo Finance symbol)",
    value="^BSESN",
    help="Enter any valid Yahoo Finance ticker. Examples: ^BSESN (Sensex), RELIANCE.NS, TCS.NS"
)

# Date range
st.sidebar.subheader("Date Range")
period = st.sidebar.selectbox(
    "Data Period",
    ["1y", "2y", "3y", "4y", "5y"],
    index=2,
    help="Amount of historical data to use"
)

# Model parameters
st.sidebar.subheader("Model Parameters")
n_estimators = st.sidebar.slider(
    "Number of Trees",
    min_value=50,
    max_value=300,
    value=100,
    step=50
)

# ==========================================================
# CACHING - Load Data
# ==========================================================

@st.cache_data(ttl=3600)
def load_stock_data(ticker, period):
    """Download stock data from Yahoo Finance"""
    try:
        df = yf.download(ticker, period=period, progress=False)
        return df
    except Exception as e:
        st.error(f"Failed to download data: {e}")
        return None

@st.cache_resource
def get_model():
    """Create and return a RandomForest classifier"""
    return RandomForestClassifier(
        n_estimators=100,
        class_weight='balanced',
        random_state=42
    )

# ==========================================================
# MAIN APP LOGIC
# ==========================================================

# Load data
with st.spinner(f"Loading data for {ticker}..."):
    df = load_stock_data(ticker, period)

if df is None or df.empty:
    st.error("Could not load data. Please check the ticker symbol.")
    st.stop()

# ==========================================================
# FIX: FLATTEN MULTIINDEX COLUMNS IF THEY EXIST
# ==========================================================

# This is the critical fix for the price_ma_ratio error
if isinstance(df.columns, pd.MultiIndex):
    df.columns = ['_'.join(col).strip() for col in df.columns.values]
    df = df.rename(columns={
        'Open_^BSESN': 'Open',
        'High_^BSESN': 'High',
        'Low_^BSESN': 'Low',
        'Close_^BSESN': 'Close',
        'Adj Close_^BSESN': 'Adj Close',
        'Volume_^BSESN': 'Volume'
    })

# ==========================================================
# DISPLAY BASIC INFO
# ==========================================================

st.subheader(f"📊 Data for {ticker}")

# Extract scalar values safely
latest_close = df['Close'].iloc[-1]
if isinstance(latest_close, pd.Series):
    latest_close = latest_close.iloc[0]

col1, col2, col3, col4 = st.columns(4)
col1.metric("Start Date", df.index[0].strftime("%Y-%m-%d"))
col2.metric("End Date", df.index[-1].strftime("%Y-%m-%d"))
col3.metric("Days", len(df))
col4.metric("Latest Close", f"${latest_close:.2f}")

# ==========================================================
# FEATURE ENGINEERING
# ==========================================================

df2 = df.copy()
df2['daily_return'] = df2['Close'].pct_change() * 100
df2['ma_7'] = df2['Close'].rolling(7).mean()
df2['ma_30'] = df2['Close'].rolling(30).mean()
df2['volatility'] = df2['daily_return'].rolling(7).std()

# FIX: Use .squeeze() to convert DataFrame to Series
df2['price_ma_ratio'] = (df2['Close'] / df2['ma_7']).squeeze()

df2['day_of_week'] = df2.index.dayofweek
df2['target'] = (df2['Close'].shift(-1) > df2['Close']).astype(int)
df2 = df2.dropna()

# ==========================================================
# TRAIN MODEL
# ==========================================================

if len(df2) < 100:
    st.warning("Not enough data for training. Please select a longer period.")
    st.stop()

# Define features and target
features = ['Close', 'Volume', 'daily_return', 'ma_7', 'ma_30', 'volatility', 'price_ma_ratio', 'day_of_week']
X = df2[features]
y = df2['target']

# Split by time (80/20)
split_idx = int(0.8 * len(X))
X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]

# Create and train pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', RandomForestClassifier(
        n_estimators=n_estimators,
        class_weight='balanced',
        random_state=42
    ))
])

pipeline.fit(X_train, y_train)

# ==========================================================
# EVALUATION
# ==========================================================

y_pred = pipeline.predict(X_test)
y_proba = pipeline.predict_proba(X_test)[:, 1]

# Display metrics
st.subheader("📊 Model Performance")
col1, col2, col3, col4, col5 = st.columns(5)
col1.metric("Accuracy", f"{accuracy_score(y_test, y_pred):.2%}")
col2.metric("Precision", f"{precision_score(y_test, y_pred):.2%}")
col3.metric("Recall", f"{recall_score(y_test, y_pred):.2%}")
col4.metric("F1 Score", f"{f1_score(y_test, y_pred):.2%}")
col5.metric("AUC", f"{roc_auc_score(y_test, y_proba):.3f}")

# ==========================================================
# PREDICTION FOR TOMORROW
# ==========================================================

st.subheader("🔮 Tomorrow's Prediction")

# Get latest data point
latest = df2.iloc[-1:][features]

# Make prediction
prediction = pipeline.predict(latest)[0]
probability = pipeline.predict_proba(latest)[0][1]

# Display prediction
col1, col2 = st.columns(2)

if prediction == 1:
    col1.success("### 📈 PREDICTION: **UP**")
else:
    col1.error("### 📉 PREDICTION: **DOWN**")

col2.metric("Confidence", f"{probability:.2%}")

# ==========================================================
# VISUALIZATIONS
# ==========================================================

st.subheader("📉 Data Visualizations")

# Price trend
fig_price = px.line(
    df,
    x=df.index,
    y='Close',
    title=f'Price Trend for {ticker}',
    labels={'Close': 'Price ($)', 'index': 'Date'}
)
st.plotly_chart(fig_price, use_container_width=True)

# Distribution of returns
fig_returns = px.histogram(
    df2,
    x='daily_return',
    nbins=50,
    title='Distribution of Daily Returns',
    labels={'daily_return': 'Daily Return (%)'}
)
st.plotly_chart(fig_returns, use_container_width=True)

# ==========================================================
# DATA TABLE
# ==========================================================

with st.expander("View Raw Data"):
    st.dataframe(df.tail(20))

st.markdown("---")
st.caption("Built with Streamlit using historical data from Yahoo Finance")
```

### **PROMPT ENGINEERING**

Prompt Engineering is the practice of designing the input we give to an LLM to get the output we want.

Its like creating instructions for an intelligent but literal assistant

A well crafted prompt uses these 4 key components:

**Instruction** :  What we want the model to do. Like Classify, Summarive, Plan

**Context** : Any info that helps the model to understand the situtation

**Input Data** : The Specific text that you want to use

**Ouput Format** : How we want the response structured, as a paragraph, or bullet points or JSON format or python program

### **RAG**

**R**etrieval **A**ugmmented **G**eneration

A RAG combines an LLM with a retrieval system. The model uses its intelligence to read and search our provided data and uses it as context to provide answers.
This ensures that answers come from a specific document.

There are 2 phases in RAG :

Indexing Phase :

- Load Document
- Split it into chunks
- Create Embedings by converting data to numerical vector to capture meaning
- Store in a vector db

Retrieval Phase :

- Embedding : Convert user query to a vector
- Retrieve Relevant Chunks : find the document chunks with the most similar vectors
- Generate Answer : Pass the retrieved chunks and the question to LLM

```python
#import libraries

from langchain_text_splitters import RecursiveCharacterTextSplitter
from sentence_transformers import SentenceTransformer
import chromadb
import tempfile
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#Creating a Sample Document

sample_text = """
Artificial Intelligence (AI) is transforming the world.

Machine Learning is a subset of AI that enables systems to learn from data without being explicitly programmed.

Deep Learning is a further subset of Machine Learning that uses neural networks with many layers.

Natural Language Processing (NLP) is a branch of AI that deals with the interaction between computers and humans using natural language.

Retrieval-Augmented Generation (RAG) is a technique that combines information retrieval with LLMs to produce more accurate and grounded responses.

Large Language Models (LLMs) like Gemini, GPT, and Claude are trained on massive datasets and can generate human-like text.

The key advantage of RAG is that it allows LLMs to access external knowledge, reducing hallucinations and enabling answers based on specific documents.

RAG works by first retrieving relevant documents from a vector database, then using those documents as context for the LLM to generate an answer.

This makes RAG particularly useful for enterprise applications like document Q&A, customer support, and knowledge management.

RAG was introduced in a 2020 paper by Lewis et al. and has since become one of the most important techniques in AI.

The RAG pipeline has two phases: the indexing phase (load, split, embed, store) and the query phase (embed query, retrieve, generate).
"""

#Save this to a temporary file

with tempfile.NamedTemporaryFile(mode='w',suffix='.txt',delete=False) as f:
  f.write(sample_text)
  sample_file_path = f.name

```

```python
#Load Document for RAG

with open(sample_file_path, 'r')as f:
  loaded_text = f.read()

print(f"Loaded Document with {len(loaded_text)} characters")
```

```python
#Split sample doc into chunks

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 300,
    chunk_overlap = 50,
    length_function = len
)

chunks = text_splitter.split_text(loaded_text)

print(f"doc split into : {len(chunks)} chunks")

print("Sample Chunk")
print(chunks[0][:200] + "....")
```

```python
# Creating Embeddings and Vector Store

embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
#Initialize the vectordb (chromadb)

chroma_client = chromadb.Client()
```

```python
#Create a new document collection

collection = chroma_client.create_collection(
    name = 'rag_demo',
    metadata = {'hnsw:space' : "cosine"}
)
```

```python
#Create embeddings

embeddings = embedding_model.encode(chunks).tolist()
```

```python
#Store these into a chromadb

for i,(text, embedding) in enumerate(zip(chunks,embeddings)):
  collection.add(
      ids = [f'chunk_{i}'],
      embeddings = [embedding],
      metadatas = [{'source':'sample_doc.txt'}],
      documents = [text]
  )

print(f"Stored {len(chunks)} embeddings in Chroma DB")
```

```python
# Adding RAG pipeline functions to this

print("Building a RAG pipeline...")

def search_similar_chunks(query,k = 3):
  """Retrieve the top k-most similar chunks for any question/query"""
  query_embedding = embedding_model.encode([query]).tolist()
  results = collection.query(
      query_embeddings = query_embedding,
      n_results = k
  )
  documents = results['documents'][0] if results['documents'] else []
  return documents

def rag_query(query, k = 3):
  """Complete RAG pipeline : retrieve -> augment -> generate."""
  retrieved_docs = search_similar_chunks(query,k)

  if not retrieved_docs:
    return {'query' : query, 'retrieved_chunks': [], 'answer' : "No relevant documents found"}

  context = "\n\n---\n\n".join(retrieved_docs)

  prompt = f"""
You are a helpful assistant that answers questions based ONLY on the provided context.
 If the context does not contain the answer, just say "I dont have enough info for the answer."

 Context : {context}

 Question : {query}

 Answer :
 """

  try:
    co = cohere.Client(os.environ.get('COHERE_API_KEY', "YOUR_API_KEY"))
    response = co.chat(
        model = COHERE_MODEL,
        message = prompt,
        temperature = 0.2
    )
    answer = response.text
  except Exception as e:
    answer = f'Error generating answer : {e}'

  return {'query' : query, 'retrieved_chunks':retrieved_docs, 'answer' : answer}


print("RAG Pipeline Created.")
```

```python
# Testing the RAG pipeline

print("Testing the RAG Pipeline........")

def display_rag_result(result):
  """Printing the RAG Result Nicely"""
  print("=" * 70)
  print(f'Question: {result['query']}')
  print("=" * 70)
  print("Retrieved Context")
  for i,doc in enumerate(result['retrieved_chunks']):
    print(f"\n [{i+1}] {doc[:150]}")
  print("=" * 70)
  print(f"Answer : {result['answer']}")
  print("=" * 70)

test_queries = [
    "What is a RAG?",
    "What is the difference between ML and DL?",
    "What is NLP?",
    "What are the key advantages of RAG?",
    "Who introduced RAG and when?"
]

for query in test_queries:
  result = rag_query(query)
  display_rag_result(result)
```
